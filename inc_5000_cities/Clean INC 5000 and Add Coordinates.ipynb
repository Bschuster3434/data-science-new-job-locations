```python
##Haversine's Formula
##Thanks to Michael Dunn
##http://stackoverflow.com/questions/4913349/haversine-formula-in-python-bearing-and-distance-between-two-gps-points

from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 

    # 3963.2 miles is the radius of the Earth
    miles = 3963.1676 * c
    return miles
```

```python
###Next section will be the function that counts
###how many cities are within r miles of a city of interest
###I am interested in getting as much data as possible, so 
###I'm considering doing every city in the list, over 1,000,000 calcs!

def grab_business_distance_information(city_state_coors, df):
    num_businesses_in_r_range = {}
    for city_code in city_state_coors.iterkeys():
        test_city = city_code
        base_coors = city_state_coors[test_city]
        distances = []
        for i in range(len(df)):
            next_city_lat = df['city_lat'].ix[i]
            next_city_lon = df['city_lon'].ix[i]
            distance_in_miles = haversine(base_coors['lon'], base_coors['lat'], next_city_lon, next_city_lat)
            distances.append(distance_in_miles)
        cities_in_range = 0
        r = 30 #miles
        for i in distances:
            if i <= r:
                cities_in_range += 1
        num_businesses_in_r_range[test_city] = cities_in_range
    return num_businesses_in_r_range

num_businesses_in_r_range = grab_business_distance_information(city_state_coors, df)
```
