```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import re
import matplotlib.pyplot as plt
```

```python
ds_jobs_file = 'data_science_jobs_info.csv'
population = 'us_population.csv'
```

```python
df_jobs = DataFrame.from_csv(ds_jobs_file, index_col = None)
df_pop = DataFrame.from_csv(population, index_col= None)
df_pop = df_pop.dropna()
```

```python
df_jobs.location.ix[17] = 'New York NY'
```

```python
df_jobs
```

```python
##Count of Relevant Columns
def relevant_content(value):
    value = value.lower()
    value = re.sub('/', ' ', value)
    value = re.sub('-', ' ', value)
    titles = ['data scientist', 'data science']
    relevant = 0
    for title in titles:
        if title in value:
            relevant = 1
            break
    return relevant

for next_title in df_jobs.columns[-6:]:
    next_name = next_title + '_count'
    df_jobs[next_name] = df_jobs[next_title].map(relevant_content)
```

```python
df_jobs['total_relevant_count'] = df_jobs.first_result_title_count + df_jobs.second_result_title_count + df_jobs.third_result_title_count + df_jobs.fourth_result_title_count + df_jobs.fifth_result_title_count + df_jobs.sixth_result_title_count
```

```python
city = [ ' '.join(city.split()[:-1]) for city in df_jobs.location]
state = [ ' '.join(state.split()[-1:]) for state in df_jobs.location]
```

```python
df_jobs['city'] = city
df_jobs['state'] = state
print df_pop.state[0]
```

```python
###Creating Unique City State Codes
df_jobs['cs_code'] = df_jobs['city'] +  [s[0] for s in df_jobs['state']]
df_pop['cs_code'] = df_pop['city'] + [s[0] for s in df_pop['state']]
```

```python
df_pop.cs_code
```

```python
df = df_jobs.merge(df_pop, how='inner', left_on = df_jobs.cs_code, right_on = df_pop.cs_code)
```

```python
df_pop.head() 
```

```python
df['jobs_per_100000'] =  df.total_results.astype(float) / df.est_pop_2013 * 100000
```

```python
result_df = df[['city_y', 'state_y', 'total_results', 'est_pop_2013', 'jobs_per_100000', 'total_relevant_count']]
```

```python
result_df.columns = ['city', 'state', 'results', 'population', 'jobs_per_100000', 'count_relevant_sample_jobs']
```

```python
result_df.sort(columns=['results'], ascending=False)
```

```python
total_jobs = 30071
result_df['percent_of_total'] = result_df.results.astype(float) / total_jobs
```

```python
result_df.head()
```

```python
fig, ax = plt.subplots()
rect = plt.bar(result_df.index, result_df.percent_of_total)
plt.xlabel('City_Id')
plt.ylabel('Total Results')
plt.show()
```
